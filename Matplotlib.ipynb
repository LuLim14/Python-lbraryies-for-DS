#Load library

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd 
```

### Basic Graph

```python
#построение прямой по точкам
x = [1, 2, 3]
y = [2, 4, 6]

#Resize your Graph
plt.figure(figsize=(5, 3), dpi=150) #fisize.firts - растяжение изображения по OX, figszie.second - растяжение по оси OY


x1 = [4, 6, 9]
y1 = [12, 18, 27]

plt.plot(x, y, color='red', linestyle='--', marker='.', markersize=10, markeredgecolor='blue') #можно работать с маркерами
plt.plot(x1, y1, color='blue')

plt.legend(['2x', '3x']) #задание легенды визуализации
plt.title('My firts title',color='green', fontdict={'fontsize': 20}) #название графика (fontdict позволяет задавать доп параметры, fontsize размер текста)
plt.xlabel('x') #задаем название оси x
plt.ylabel('y') #задаем название оси y

#plt.xticks([1, 2, 3, 5]) #указываем, какие координаты по x хотим видеть
#plt.yticks([2, 4, 6, 10]) # аналогично по y
```

```python
#Use shorthand notation
#fmt = '[marker][line][color]'

plt.plot(x, y, 'b^--')
plt.legend(['2x'])
```

```python
#Line Number Two
x2 = np.arange(0, 4, 0.5) #кажется, какое-то распределение (нормальное?)
plt.plot(x2, x2**2, 'red')
plt.legend(['x^2'])
```

### Bar Chart

```python
labels = ['A', 'B', 'C']
values = [1, 4, 2]
plt.bar(labels, values) #ступенчатая  диаграма
plt.figure(figsize=(3, 5), dpi=300)
```

```python
labels = ['A', 'B', 'C']
values = [1, 4, 2]
bars = plt.bar(labels, values) 

bars[0].set_hatch('/')
bars[1].set_hatch('O')

plt.figure(figsize=(3, 5), dpi=300)
```
