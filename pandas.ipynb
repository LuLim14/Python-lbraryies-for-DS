# Making changes to the data

```python
df.head(5)
```

```python
df['Total'] = df['HP'] + df['Attack'] + df['Defense'] + df['Sp. Atk'] + df['Sp. Def'] + df['Speed'] #добавление нового столбца
df.head(5)
```

```python
df = df.drop(columns=['Total']) #удаление столбцов
df.head(5)
```

```python
df['Total'] = df.iloc[: , 4:10].sum(axis=1) #sum(axis=1) - счиатем сумму в строке (параллельно оси x), иначе параллельно оси y
df.head(5)
```

```python
cols = list(df.columns.values)
df = df[cols[0:4] + [cols[-1]] + cols[4:12]] #можно так переставлять столбцы как удобно
df.head(5)
```

#Aggregate (Groupby)

```python
df.groupby(['Type 1']).mean().sort_values('HP', ascending=False) #группируем по столбцу Type 1 и считаем статистики по такой группировке по всем столбцам, а потом сортим
#по стобцу Defense
```

```python
df.groupby(['Type 1']).sum() #группируем по Type 1 и суммирумем в каждой группе
```

```python
df.groupby(['Type 1']).count() #функция count в каждой группе
```

```python
# как можно выделить отдельный столбец для кол-ва по группе
df['count'] = 1
#df.groupby(['Type 1']).count()
df.groupby(['Type 1', 'Type 2']).count()['count']
```
